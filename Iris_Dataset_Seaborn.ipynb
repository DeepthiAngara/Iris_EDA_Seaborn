## 1. Distribution plots
 1. distplot
 2. joinplot

### 1.1 distplot

```python
iris.columns
```

```python
# Select only columns of int and float types
num_cols = iris.select_dtypes(include=['int', 'float']).columns

# Iterate over selected columns
for col in num_cols:
    sns.distplot(iris[col], label=col)

# Add lables and legend to the plot
plt.xlabel('Numerical Columns')
plt.ylabel('Distribution')
plt.legend()
plt.show()
```

## 2. Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

1. countplot 
2. bar plot 
3. boxplot
4. violinplot

### 2.1. countplot

```python
sns.countplot(x='species', data=iris)
```

### 2.2 Bar plot

```python
# Create a figure with four subplots arranged in a 2x2 grid
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

# Select only columns of int and float types
num_cols = iris.select_dtypes(include=['int', 'float']).columns

# Iterate over selected columns and create a bar plot in each subplot
for i, col in enumerate(num_cols):
    sns.barplot(x='species', y=col, data=iris, ax=axes[i//2, i%2])

# Add some space between the subplots
fig.tight_layout()

# Show the plot
plt.show()
```

## 2.3 Boxplot

```python
sns.boxplot(data=iris)
```

```python
# Create a figure with four subplots arranged in a 2x2 grid
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

# Select only columns of int and float types
num_cols = iris.select_dtypes(include=['int', 'float']).columns

# Iterate over selected columns and create a bar plot in each subplot
for i, col in enumerate(num_cols):
    sns.boxplot(x='species', y=col, data=iris, ax=axes[i//2, i%2])

# Add some space between the subplots
fig.tight_layout()

# Show the plot
plt.show()
```

## 2.4 Violin Plot

```python
# Create a figure with four subplots arranged in a 2x2 grid
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

# Select only columns of int and float types
num_cols = iris.select_dtypes(include=['int', 'float']).columns

# Iterate over selected columns and create a bar plot in each subplot
for i, col in enumerate(num_cols):
    sns.violinplot(x='species', y=col, data=iris, ax=axes[i//2, i%2])

# Add some space between the subplots
fig.tight_layout()

# Show the plot
plt.show()
```
